2.  Explain the difference between mutable and immutable data types with examples.
    - a. Mutable data type =
          - These are the data type whose value can be changed after creation.
          - They retain same memory location even after creation.
          - They are memory efficient, as one can alter it.
          - for eg. list, dictionary,sets
    - Immutable  data type =
          - These are the data types whose values cannot be changed after creation.
          - it create new memory location after creation.
          - In this data remain consistent and fixed.
          - for eg. strings, tuples.

8. What is a hash table, and how does it relate to dictionaries in Python?
    - Hash tables are a type of data structure in which the address or the index value of the data element is generated from a hash function. Hash table stores key-value pairs but the key is generated through a hashing function.
    - In Python, the Dictionary data types represent the implementation of hash tables. The keys of the dictionary are hashable i.e. the are generated by hashing function which generates unique result for each unique value supplied to the hash function.
